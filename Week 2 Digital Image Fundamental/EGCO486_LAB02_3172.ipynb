```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros([256, 256], np.uint8)
plt.imshow(image, cmap = 'gray')
plt.show()
```

```python
image = np.zeros([256, 256], np.uint8)
for i in range (0, 256):
  image[:, i] += i
plt.title('image of rmap 256 gray level')
plt.imshow(image, cmap = 'gray')
plt.show()
```

```python
moon_reshape = []
moon_reshape.append(moon)

for i in range(0, 5):
  print(int(moon_reshape[i].shape[0]/2))
  moon_reshape.append(cv2.resize(moon, (int(moon_reshape[i].shape[0]/2), int(moon_reshape[i].shape[1]/2))))
```

```python
plt.imshow(moon_reshape[5])
plt.show()
```

```python
for i in range(0, 6):
    plt.subplot(2,3,i+1),plt.imshow(moon_reshape[i]),plt.title(str(moon_reshape[i].shape[0]) + 'x' + str(moon_reshape[i].shape[1]))
    plt.xticks([]), plt.yticks([])
plt.show()
```
