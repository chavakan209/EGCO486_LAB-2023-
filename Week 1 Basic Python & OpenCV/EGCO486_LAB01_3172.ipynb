```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
image = np.zeros([8, 8], np.uint8)
image[:, 2:6] = 255
image[2, 2:4] = 0
image[5, 2:4] = 0
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# Using RGB
image = np.ones([8, 8, 3], np.uint8)

image *= 255

image[:, 2:6, 0] = 255
image[2, 2:4, 0] = 255
image[5, 2:4, 0] = 255

image[:, 2:6, 1] = 0
image[2, 2:4, 1] = 255
image[5, 2:4, 1] = 255

image[:, 2:6, 2] = 0
image[2, 2:4, 2] = 255
image[5, 2:4, 2] = 255

plt.imshow(image)
plt.show()
```
